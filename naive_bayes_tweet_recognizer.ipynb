# **Naive Bayes classifier using Multinomial Naive Bayes made to predict who made the tweet? Was it Kanye, Musk or Trump?**

### Data cleaning

```python
# importing packages for reading data
import pandas as pd
```

```python
# checking the layout of first .csv
kanye = pd.read_csv('kanye_tweets.csv')
kanye.head()
```

```python
# cleaning data
kanye = kanye.drop(columns = ['Unnamed: 0'], axis = 1)
kanye.head()
```

```python
# adding a separate column for our target class
kanye['target'] = 'kanye'
kanye.head()
```

```python
# checking the layout of second .csv
musk = pd.read_csv('musk_tweets.csv')
musk.head()
```

```python
# dropping unnecessary columns and adding a separate column for our target class
musk = musk['fullText']
musk = pd.DataFrame({'Tweet': musk.values, 'target': 'musk'})
musk.head()
```

```python
# loading our final .csv
trump = pd.read_csv('trump_tweets.csv')
trump.head()
```

```python
# dropping unnecessary columns and adding a separate column for our target class
trump = trump['status_text']
trump = pd.DataFrame({'Tweet': trump.values, 'target': 'trump'})
trump.head()
```

```python
# mergin all three .csv files into one using pd.concat
data = pd.concat([kanye, musk, trump], ignore_index = True)
data.head()
```

```python
# looks like we have some NaN values we need to drop
data.isna().sum()
```

```python
data = data.dropna()
```

```python
# our final dataset seems to be too 'Musk-heavy' but it should be alright
data.groupby('target').describe()
```

### Training our model

```python
# splitting our dataset into features and target variables
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.Tweet, data.target, test_size = 0.20)
```

### Initializing our CountVectorizer

```python
# CountVectorizer will convert text to a matrix of token counts and try to learn vocabulary
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_test_count = v.transform(X_test)
```

```python
# fitting our data into our model
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB().fit(X_train_count, y_train)
```

```python
# model gives out 95% confidence
model.score(X_test_count, y_test)
```

```python
# creating test tweets one for each to run with our model
tweets = [
    'My ideas scare people who are trapped in tradition.',
    'Reality is just a low-res simulation, probably.',
    'When America wins, everybody wins — remember that!'
  ]
```

```python
# our model correctly predicts who made which tweet
tweets_count = v.transform(tweets)
model.predict(tweets_count)
```
